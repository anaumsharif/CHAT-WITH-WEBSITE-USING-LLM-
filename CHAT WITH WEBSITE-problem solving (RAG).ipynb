```python
from langchain.document_loaders import WebBaseLoader
from langchain.text_splitter import RecursiveCharacterTextSplitter
from langchain.embeddings import HuggingFaceInferenceAPIEmbeddings
from langchain.vectorstores import Chroma

from langchain import HuggingFaceHub
from langchain.chains import RetrievalQA
```

```python
import os
from getpass import getpass
HF_token = getpass()
```

```python
os.environ['HUGGINGFACEHUB_API_TOKEN']=HF_token
```

```python
URL ="https://privatewriting.net/blog/problem-solving-skills"
```

```python
data =WebBaseLoader(URL)
```

```python
content = data.load()
```

```python
print(content)
```

# CHUNKING -Text Splitters

```python
text_splitter = RecursiveCharacterTextSplitter(chunk_size=256,chunk_overlap=0)
```

```python
chunking =text_splitter.split_documents(content)
```

```python
len(chunking)
```

```python
chunking[2]
```

```python
chunking[3]
```

```python
chunking[4]
```

# EMBEDDING MODEL

```python
embeddings =HuggingFaceInferenceAPIEmbeddings(
    api_key = HF_token ,model_name='BAAI/bge-base-en-v1.5'
 )
```

```python
vectorstore = Chroma.from_documents(chunking,embeddings)
```

```python
print(embeddings)
```

```python
model = HuggingFaceHub(repo_id="HuggingFaceH4/zephyr-7b-alpha",
                       model_kwargs={"temperature":0.5,"max_new_tokens":512,"max_length":64})
```

```python
retriver = vectorstore.as_retriever(search_type="mmr",search_kwargs={"k":1})
```

```python
query = "What works like brainstorming?"
```

```python
docs_rel = retriver.get_relevant_documents(query)
```

## Step-2 : Augment --- Prompt

```python
query = "What  to effectively solve problems?"
```

```python
prompt = f"""
<|system|>
You are an AI generated Prompy assistant
Provide an output thats is factully correct and authentic
</s>
<|user|>
{query}
</s>
<|assistant|>
"""
```

## Step-3 : Generation



```python
qa = RetrievalQA.from_chain_type(llm=model,retriever=retriver,chain_type="stuff")
```

```python
response = qa(prompt)
```
